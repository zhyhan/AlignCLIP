```python
import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['Domain', 'Domain+Class']
conditions = ['AAL: O / SLS: O', 'AAL: X / SLS: O', 'AAL: O / SLS: X', 'AAL: X / SLS: X']
accuracy_values = {
    'Domain': [86.9, 86.7, 86.6, 86.1],
    'Domain+Class': [84, 83.5, 83.3, 82.6]
}
colors = ['#76c7a1', '#b8d18c', '#ffa270', '#ff8374']

# Initialize the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Bar width
bar_width = 0.1

# X-axis position for each category
x = np.arange(len(categories))

# Find the minimum value for each category
min_values = {cat: min(accuracy_values[cat]) for cat in categories}

# Plot bars and annotate differences
for i, condition in enumerate(conditions):
    for j, cat in enumerate(categories):
        y_value = accuracy_values[cat][i]
        diff = y_value - min_values[cat]

        # Plotting the bar
        bar = ax.bar(x[j] + i*bar_width, y_value, width=bar_width, color=colors[i], label=condition if j == 0 else "")

        # Annotating the difference
        ax.annotate(f'+{diff:.1f}', 
                    (bar[0].get_x() + bar[0].get_width() / 2, y_value), 
                    textcoords="offset points", 
                    xytext=(0,10), 
                    ha='center',
                    fontsize=20)

# Set y-axis label and range
ax.set_ylabel('Accuracy (%)', fontsize=30)

ax.set_ylim([82, 88])
ax.set_yticklabels(y_ticks, fontsize=30)
y_ticks = np.arange(82, 88, 2)
ax.set_yticks(y_ticks)
#ax.set_yticklabels(y_ticks, fontsize=30)
# y_ticks = np.arange(82, 87, 2)
# ax.set_yticks(y_ticks)
# ax.set_yticklabels(y_ticks, fontsize=10)

# Set x-axis ticks
ax.set_xticks(x + bar_width * 1.5)
ax.set_xticklabels(categories, fontsize=30)

# Add a legend
ax.legend(loc='best', fontsize=30)

plt.savefig('analysis_ablation_revise.pdf')
# Show the plot
plt.show()
```
